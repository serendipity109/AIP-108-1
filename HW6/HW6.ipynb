```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# Gassian filter
def gaussian_smooth(size):
    x, y = np.mgrid[-(size-2):(size-1), -(size-2):(size-1)]
    gaussian_kernel = np.multiply(1/(2*pi*(9**2)), np.exp(-(x**2+y**2)/(2*(9**2))))
    gaussian_kernel = gaussian_kernel / gaussian_kernel.sum()
    return gaussian_kernel
```

```python
blur = convolve(img, gaussian_smooth(5))
x, y = xyfilter(5)
a = convolve(blur, x)
b = convolve(blur, y)
c = np.sqrt(np.add(np.square(a), np.square(b)))
c = c / c.max() * 255
```

```python
theta = np.arctan2(b, a)
def non_max_suppression(img, D):
    M, N = img.shape
    Z = np.zeros((M,N), dtype=np.int32)
    angle = D * 180. / np.pi
    angle[angle < 0] += 180

    
    for i in range(1,M-1):
        for j in range(1,N-1):
            try:
                q = 255
                r = 255
                
               #angle 0
                if (0 <= angle[i,j] < 22.5) or (157.5 <= angle[i,j] <= 180):
                    q = img[i, j+1]
                    r = img[i, j-1]
                #angle 45
                elif (22.5 <= angle[i,j] < 67.5):
                    q = img[i+1, j-1]
                    r = img[i-1, j+1]
                #angle 90
                elif (67.5 <= angle[i,j] < 112.5):
                    q = img[i+1, j]
                    r = img[i-1, j]
                #angle 135
                elif (112.5 <= angle[i,j] < 157.5):
                    q = img[i-1, j-1]
                    r = img[i+1, j+1]

                if (img[i,j] >= q) and (img[i,j] >= r):
                    Z[i,j] = img[i,j]
                else:
                    Z[i,j] = 0

            except IndexError as e:
                pass
    
    return Z
z = non_max_suppression(c, theta)
```
